```python
#  
import numpy as np
import matplotlib.pyplot as plt

y1 = lambda x: (-8*x+42)/21
y2 = lambda x: (-8*x-168)/-42
y3 = lambda x: (16*x-168)/21
z = lambda x: (8*x)/-21

fig = plt.subplots(figsize=(14, 7))

x = np.linspace(0, 25,100)

plt.plot(x, y1(x), label='-8x1 - 21x2 = -42')
plt.plot(x, y2(x), label='-8x1 + 42x2 = 168')
plt.plot(x, y3(x), label='16x1 - 21x2 = 168')
plt.plot(x, z(x), label='8x1 + 21x2')

plt.title(' ', fontsize=15)

plt.legend()
plt.grid(True)
#plt.text(15, 4, 'grow up!')

plt.show()
```
