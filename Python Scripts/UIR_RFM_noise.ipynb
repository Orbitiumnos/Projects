# Граббс

```python
from scipy.stats import *
```

```python
alpha=0.05
```

```python
def critical_point(alpha, N):
    st_t=abs(t.ppf(alpha/(2*N), N-2))
    return ((N-1)/(np.sqrt(N)))*np.sqrt((st_t**2)/(N-2+st_t**2))
```

```python
df4=df2.copy()
```

```python
for j in range(df4.shape[1]):
    feat_test=df4.ix[:, j].values
    _, p = normaltest(feat_test)
    if p<alpha:
        print(j)
        continue
    while True:
        column=df4.ix[:, j].values
        mean=column.mean()
        std=column.std()
        col_m=abs(column-mean)
        outlier=np.argmax(col_m)
        G=max(col_m)/std
        if G>critical_point(alpha, column.shape[0]):
            df4.drop(df4.index[outlier], inplace=True)
        else:
            break
```

# Пирс

```python
import scipy.special
```

```python
def peirce_dev(N, n, m):
    """
    Name:     peirce_dev
    Input:  - int, total number of observations (N)
             - int, number of outliers to be removed (n)
             - int, number of model unknowns (m)
    Output:   float, squared error threshold (x2)
    Features: Returns the squared threshold error deviation for outlier 
             identification using Peirce's criterion based on Gould's
             methodology
    """
    # Assign floats to input variables:
    N = float(N)
    n = float(n)
    m = float(m)
    #
    # Check number of observations:
    if N > 1:
        # Calculate Q (Nth root of Gould's equation B):
        Q = (n**(n/N)*(N - n)**((N - n)/N))/N
        #
        # Initialize R values (as floats)
        Rnew = 1.0  
        Rold = 0.0  # <- Necessary to prompt while loop
        #
        # Start iteration to converge on R:
        while ( abs(Rnew - Rold) > (N*2.0e-16) ):
            # Calculate Lamda 
            # (1/(N-n)th root of Gould's equation A'):
            ldiv = Rnew**n
            if ldiv == 0:
                ldiv = 1.0e-6
            Lamda = ((Q**N)/(ldiv))**(1.0/(N - n))
            #
            # Calculate x-squared (Gould's equation C):
            x2 = 1.0 + (N - m - n)/n*(1.0 - Lamda**2.0)
            #
            # If x2 goes negative, return 0:
            if x2 < 0:
                x2 = 0.0
                Rold = Rnew
            else:
                # Use x-squared to update R (Gould's equation D):
                Rold = Rnew
                Rnew = (
                   np.exp((x2 - 1)/2.0)*
                   scipy.special.erfc(np.sqrt(x2)/np.sqrt(2.0))
                   )
    else:
        x2 = 0.0
    return x2
```

# Межквартильные расстояния

```python
low_lev=25
up_lev=75
coef=1.5
```

```python
df4=df2.copy()
```

```python
for j in range(df4.shape[1]):
    column=df4.ix[:, j].values
    df_v_i=df4.ix[:, j]
    low=np.percentile(column, low_lev)
    up=np.percentile(column, up_lev)
    low_lim=low-coef*(up-low)
    up_lim=up+coef*(up-low)
    for i in range(column.shape[0]):
        if column[i]<low_lim or column[i]>up_lim:
            df4.drop(df_v_i.index[i], inplace=True)
```
